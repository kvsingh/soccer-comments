```python
import pickle
from config import *
import numpy as np
import matplotlib.pyplot as plt

comment_tokens = pickle.load(open("comment_tokens.p", "rb"))

for subreddit in subreddits:
    #number of tokens in each comment
    this_lens = []
    this_comment_tokens = comment_tokens[subreddit]
    for comment in this_comment_tokens:
        this_lens.append(len(comment))
    print subreddit, np.mean(this_lens), np.median(this_lens), np.std(this_lens)
    plt.hist(this_lens)
    plt.show()
```
